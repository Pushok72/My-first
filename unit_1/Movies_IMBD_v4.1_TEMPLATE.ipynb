# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
data2 = data[['original_title', 'production_companies']]
data2 = pd.DataFrame(data2)
pd.set_option('display.max_colwidth', 200)
data2['lenght'] = data2.original_title.apply(lambda x: len(x))
data2.sort_values('lenght', ascending = False)
```

# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

```python
data2 = data[['overview', 'production_companies']]
data2 = pd.DataFrame(data2)
pd.set_option('display.max_colwidth', 200)
data2['len'] = data2.overview.str.split(' ').apply(lambda x:len(x))
data2.sort_values('len')
```

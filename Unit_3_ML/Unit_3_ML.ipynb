```python
import pandas as pd
import numpy as np
```

```python
"""

Первичная обработка

"""

users = pd.read_csv('users.csv', encoding = 'koi8_r', sep = '\t')
users.columns = ['user', 'mail', 'geo']
users.user = users.user.apply(lambda x:  str(x).lower())
display(users.head())

log = pd.read_csv('log.csv', header = None)
display(log)

log.columns = ['user', 'time', 'bet', 'win']
log.user = log.user.apply(lambda x: x[x.find('-')+2:])
log.time = log.time.apply(lambda x: str(x[1:]) if '[' in str(x) else x)

display(log)
```

```python
log.time.isna().value_counts()
```

```python
log.time.dropna() #axis = 0 string /  axis = 1 column 
```

```python
log.bet.isna().value_counts()
```

```python
log.dropna(subset = ['bet', 'win', 'time'] ) # Define in which columns to look for missing values.
log.dropna(axis = 'columns') #Drop the columns where at least one element is missing.
log.time.dropna() #axis = 0 string /  axis = 1 column 

log.time = log.time.apply(lambda x: x.minute)
```

```python
log = log.drop_duplicates(subset = 'time')
log.loc[:, 'time'] = pd.to_datetime(log.time) # format column 'time' from string to time :)

log.time.max()
```
